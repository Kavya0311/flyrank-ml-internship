3. Quick Look at the Data

To support my choice of Lane 2, I explored the starter dataset and calculated a few summary statistics. These observations help determine whether the dataset contains enough variation to support a content refresh prioritization problem.

```python
!git clone https://github.com/Kavya0311/flyrank-ml-internship.git
```

```python
import pandas as pd

# Load the starter dataset
df = pd.read_csv("/content/flyrank-ml-internship/data/raw/content_refresh_anonymized.csv")

print("Total Pages:", len(df))

print("\nTrend Direction Counts:")
print(df["trend_direction"].value_counts())

print("\nAverage CTR:")
print(round(df["ctr"].mean(), 4))

print("\nAverage Impressions (90 Days):")
print(round(df["impressions_90d"].mean(), 2))

```

Observation

The starter dataset contains 30,000 content pages. Among them, 16,262 pages are marked with a downward trend, indicating that a substantial number of pages may require attention. The dataset also has an average CTR of 0.5107 and an average of 5,200.37 impressions over the last 90 days. These observed statistics suggest that page performance varies considerably, making content refresh prioritization a meaningful decision-support problem for machine learning.
